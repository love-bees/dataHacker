## First "real" dataset of machine learning

Until now we have used hand made datasets or those made using commands. One of the first sets of data that scientists used is the Iris Dataset. Almost all blogs or textbooks use this dataset, and we won't make an exception in using it ourselves. This set of data was made by a statistician Fisher (1936) and he also contributed to the development of the LDA algorithm. You will often find it named as the Fished Linear DIscrimnant Analysis. Fisher wanted to know if three different kinds of iris flowers could be classified based on the width and length of the petals.

image.png

In this case we will analyze 3 classes, and those are the categories of the flower iris: setosa, versicolor and virginica. On the previous picture we have illustrated the dimensions that are actually being measured. From each flower then we get 4 parameters. These dimensions are called features in machine learning, and these dimensions together are called a feature vector.


In data science, depending on the problem that's being solved, we need to know the domain expertise as well. At least one member of the data science team should be able to posess this. Here we would have to, for example, consult a biologist, while for an application to the medical field we would need doctors on the team. The following code imports the Iris dataset.

```
# Importing standard libraries

import numpy as np
import matplotlib.pyplot as plt
```

```
# Importing the Iris dataset
# In CoLab the sklearn library is preinstalled
# Inside of that library are excellent premade datasets for practice

from sklearn.datasets import load_iris
```

```
# These three commands are enough to import the iris dataset
# For now, there's no need to analyze what's stored in load_iris()
iris = load_iris()
X = iris.data
y = iris.target
```

Let's see how the dataset looks. This is a very important part in data science. The percise name is EDA - Exploratory Data Analysis.

In this step, we're trying to understand the weight of the problem, dependencies between feature vectors, if the data is all correct or if some values are missing etc.
